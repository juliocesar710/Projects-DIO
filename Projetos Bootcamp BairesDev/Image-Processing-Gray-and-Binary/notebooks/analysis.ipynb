```python
# Importar bibliotecas
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

print("Bibliotecas importadas com sucesso!")
```

```python
# Upload da imagem
from google.colab import files

print("Faa o upload da sua imagem (por exemplo, Lena):")
uploaded = files.upload()  # Abre dilogo para upload
image_path = list(uploaded.keys())[0]  # Pega o nome do arquivo carregado

print(f"Imagem carregada: {image_path}")
```

```python
# Abrir a imagem
image = Image.open(image_path)

# Mostrar a imagem original
plt.imshow(image)
plt.title("Imagem Original")
plt.axis("off")
plt.show()

print("Imagem original exibida com sucesso!")
```

```python
# Converter a imagem para tons de cinza
gray_image = image.convert("L")

# Exibir a imagem em tons de cinza
plt.imshow(gray_image, cmap="gray")
plt.title("Imagem em Tons de Cinza")
plt.axis("off")
plt.show()

print("Imagem convertida para tons de cinza com sucesso!")
```

```python
# Definir o limite para binarizao
threshold = 128

# Aplicar binarizao
binary_image = gray_image.point(lambda p: 255 if p > threshold else 0, mode='1')

# Exibir a imagem binarizada
plt.imshow(binary_image, cmap="gray")
plt.title("Imagem Binarizada (Preto e Branco)")
plt.axis("off")
plt.show()

print("Imagem binarizada com sucesso!")
```

```python
# Salvar a imagem em tons de cinza
gray_image.save("gray_image.png")
print("Imagem em tons de cinza salva como 'gray_image.png'.")

# Salvar a imagem binarizada
binary_image.save("binary_image.png")
print("Imagem binarizada salva como 'binary_image.png'.")
```

```python
# Exibir as trs verses da imagem
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Imagem original
ax[0].imshow(image)
ax[0].set_title("Imagem Original")
ax[0].axis("off")

# Imagem em tons de cinza
ax[1].imshow(gray_image, cmap="gray")
ax[1].set_title("Imagem em Tons de Cinza")
ax[1].axis("off")

# Imagem binarizada
ax[2].imshow(binary_image, cmap="gray")
ax[2].set_title("Imagem Binarizada")
ax[2].axis("off")

plt.tight_layout()
plt.show()

print("As trs verses da imagem foram exibidas para comparao.")
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
from ipywidgets import interact, IntSlider
import cv2
import matplotlib.pyplot as plt

# Funo para exibir a imagem ajustada
def display_adjusted_binary(threshold):
    _, binary_adjusted = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_BINARY)
    plt.figure(figsize=(5, 5))
    plt.imshow(binary_adjusted, cmap="gray")
    plt.title(f"Imagem Binarizada - Threshold {threshold}")
    plt.axis("off")
    plt.show()

# Criar o slider para ajuste do threshold
interact(display_adjusted_binary, threshold=IntSlider(min=0, max=255, step=1, value=127));
```
